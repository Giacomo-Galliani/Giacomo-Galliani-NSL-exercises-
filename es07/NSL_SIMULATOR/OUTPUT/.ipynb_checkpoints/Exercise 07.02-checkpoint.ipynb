per i primi 5000 step di dinamica molecolare, partendo da una configurazione cristallina a bassa entropia, la temperatura $T$ evolve nel tempo, con un graduale raffreddamento del sistema, dovuto all'aumentare dell'energia potenziale (e alla conseguente diminuzione dell'energia cintetica).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

z = np.loadtxt("temperature_beginning_MD.dat")

z.reshape(-1, 4)
x = z[:,0]
y =z[:,1]


plt.plot(x,y)
plt.grid(True)
plt.show()
```

Osserviamo dunque che per avere la temperatura all'equilibrio di $T^* = 1.1$, è necessario partire da una temperatura di $T =2.0$. 

Ora stampiamo i valori istantanei per le energie potenziali medie per particella ($U/N$), sia per l'algoritmo di dinamica molecolare, sia per montecarlo, in funzione dei passi (dopo avere equilibrato il sistema):

```python
z = np.loadtxt("potential_energy_MD.dat")

z.reshape(-1, 4)
x = z[:,0]
y1 =z[:,1]


plt.plot(x,y1)
plt.title("$U/N$ - Molecolar Dynamics")
plt.grid(True)
plt.show()
```

```python
z = np.loadtxt("potential_energy_MC.dat")

z.reshape(-1, 4)
y2 =z[:,1]


plt.plot(x,y2)
plt.title("$U/N$ - Monte-Carlo")
plt.grid(True)
plt.show()
```

Ora tracciamo le funzioni di autocorrelazione $\chi_{MD}(t)$ e $\chi_{MC}(t)$, per stimare la lunghezza dei blocchi:

```python
tm = int(x.size / 500)

# Preparo gli accumulatori
a1 = np.zeros(tm)
b1 = np.zeros(tm)
c1 = np.zeros(tm)
chi1 = np.zeros(tm)
f1 = np.arange(tm)


for j in range(tm):
    for i in range(tm - j):
        a1[j] += y1[i] * y1[i + j]
        b1[j] += y1[i]
        c1[j] += y1[i + j]


d1 = 0.0
e1 = 0.0
for i in range(tm):
    d1 += y1[i] * y1[i]
    e1 += y1[i]

# Calcolo chi(t)
for i in range(tm):
    norm = tm - i
    num = (a1[i] / norm) - (b1[i] / norm) * (c1[i] / norm)
    den = (d1 / tm) - (e1 / tm)**2
    chi1[i] = num / den 

# Plot
plt.plot(f1, chi1)
plt.grid(True)
plt.title("$\chi_{MD}(t)$")
plt.xlabel("t")
plt.ylabel("χ(t)")
plt.show()

#stessa cosa per Monte Carlo

a2 = np.zeros(tm)
b2 = np.zeros(tm)
c2 = np.zeros(tm)
chi2 = np.zeros(tm)


for j in range(tm):
    for i in range(tm - j):
        a2[j] += y2[i] * y2[i + j]
        b2[j] += y2[i]
        c2[j] += y2[i + j]


d2 = 0.0
e2 = 0.0

for i in range(tm):
    d2 += y2[i] * y2[i]
    e2 += y2[i]

for i in range(tm):
    norm = tm - i
    num = (a2[i] / norm) - (b2[i] / norm) * (c2[i] / norm)
    den = (d2 / tm) - (e2 / tm)**2
    chi2[i] = num / den if den != 0 else 0


plt.plot(f1, chi2)
plt.grid(True)
plt.title("$\chi_{MC}(t)$")
plt.xlabel("t")
plt.ylabel("χ(t)")
plt.show()
```

Osseriviamo dunque che per costruire dei blocchi statisticamente indipendenti dovremmo avere una lunghezza di circa $N=200$

Determiniamo dunque ora l'incertezza dei nostri dati al variare di L, ovvero la lunghezza dei blocchi. 

```python
m = x.size
L_list = [10, 50, 200, 500, 1000, 2000, 3000, 4000, 5000]
err = np.zeros(len(L_list))
w = np.array(L_list, dtype=float)

for idx, L in enumerate(L_list):
    N = int(m / L)
    ave = np.zeros(N)

    for i in range(N):
        start = i * L
        ave[i] = np.mean(y1[start:start + L])  # ← media corretta del blocco

    average = np.mean(ave)
    sigma = np.mean(ave ** 2) - average ** 2
    err[idx] = sigma

# Calcolo dell'errore sulla media:
err = np.sqrt(err) / np.sqrt(m / w)

# Plot
plt.plot(w, err, marker='o')
plt.grid(True)
plt.xlabel("L (dimensione del blocco)")
plt.ylabel("Errore sulla media")
plt.title("Errore in funzione della dimensione dei blocchi - MD")
plt.show()
```

Come ci attendevamo l'incertezza viene sottostimata per i blocchi più piccoli. converge rapidamente verso la reale incertezza invece per blocchi con $L>200$. Vediamo anche il grafico per il Monte Carlo: 

```python
m = x.size
L_list = [10, 50, 200, 500, 1000, 2000, 3000, 4000, 5000]
err = np.zeros(len(L_list))
w = np.array(L_list, dtype=float)

for idx, L in enumerate(L_list):
    N = int(m / L)
    ave = np.zeros(N)

    for i in range(N):
        start = i * L
        ave[i] = np.mean(y2[start:start + L])  # ← media corretta del blocco

    average = np.mean(ave)
    sigma = np.mean(ave ** 2) - average ** 2
    err[idx] = sigma

# Calcolo dell'errore sulla media:
err = np.sqrt(err) / np.sqrt(m / w)

# Plot
plt.plot(w, err, marker='o')
plt.grid(True)
plt.xlabel("L (dimensione del blocco)")
plt.ylabel("Errore sulla media")
plt.title("Errore in funzione della dimensione dei blocchi - MC")
plt.show()
```

E come previsto ha un comportamento simile.
