Estendiamo il nostro modello in modo tale che sia in grado di fittare una funzione 2D:

```python
import math
import numpy as np

#generate training input
np.random.seed(0)
x_train = np.random.uniform(-1.5, 1.5, 500)
x_valid = np.random.uniform(-1.5, 1.5, 50)
y_train = np.random.uniform(-1.5, 1.5, 500)
y_valid = np.random.uniform(-1.5, 1.5, 50)
x_valid.sort()
y_valid.sort()

z_target_x0 = np.sin(y_valid*y_valid)
z_target_y0 = np.sin(x_valid*x_valid)
z_target = np.sin(x_valid*x_valid + y_valid*y_valid)


sigma = 0 # noise standard deviation, for the moment it is absent
z_train = np.random.normal(np.sin(x_train*x_train + y_train*y_train), sigma) # actual measures from which we want to guess regression parameters
z_valid = np.random.normal(z_target , sigma)
z_valid_x0 = np.random.normal(z_target_x0, sigma)
z_valid_y0 = np.random.normal(z_target_y0 , sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, z_target_y0, label='target')
plt.scatter(x_valid, z_valid_y0, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()


plt.plot(y_valid, z_target_x0, label='target')
plt.scatter(y_valid, z_valid_x0, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

Per una funzione 2D, usando come ottimizzatore adam, abbiamo riscontrato che, a differenza del sgd, il rumore gaussiano non aiuta la convergenza dell'algoritmo

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(2, input_shape=(2,)))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(64, activation= 'relu'))
model.add(Dense(64, activation= 'relu'))
model.add(Dense(64, activation= 'relu'))
model.add(Dense(64, activation= 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data

X_train = np.stack((x_train, y_train), axis=1)
X_valid = np.stack((x_valid, y_valid), axis=1)

history = model.fit(x=X_train, y=z_train,
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(X_valid, z_valid))
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1.5, 1.5, 100)
y_predicted = np.zeros(100)

X_predicted = np.stack((x_predicted, y_predicted), axis=1)

z_predicted = model.predict(X_predicted)
plt.scatter(x_predicted, z_predicted,color='r')
plt.plot(x_valid, z_target_y0)
plt.grid(True)
plt.title('Fit vs Test data (section y=0)')
plt.show()

y_predicted = np.random.uniform(-1.5, 1.5, 100)
x_predicted = np.zeros(100)

X_predicted = np.stack((x_predicted, y_predicted), axis=1)

z_predicted = model.predict(X_predicted)
plt.scatter(y_predicted, z_predicted,color='r')
plt.plot(y_valid, z_target_x0)
plt.grid(True)
plt.title('Fit vs Test data (section x=0)')
plt.show()
```
