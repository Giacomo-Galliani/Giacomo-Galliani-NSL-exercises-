Per prima cosa, stampiamo le medie ed errori progressivi dell'integrale $\int_0^1 r dr$, meno il suo risultato noto, in funzione del numero di blocchi: (abbiamo usato $N=10^5$ lanci, e $N_{blocks} =100$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y = np.loadtxt("mean_01.txt")
err = np.loadtxt("err_01.txt")
with open("integers.txt") as file:
    riga = file.readline()
    N, L = map(int, riga.split())

x = np.arange(N)
x *= L

plt.errorbar(x,y-0.5,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

Stampiamo ora la varianza dell'integrale e i suoi errori progressivi, sempre meno il suo risultato noto, in funzione del numero di blocchi: 

```python
y = np.loadtxt("mean_02.txt")
err = np.loadtxt("err_02.txt")


plt.errorbar(x,y-1/12,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

Infine, stampiamo i risultati del test del $\chi^2$ (nell'approssimazione di Pearson), su ogni singolo blocco, avente 100 elementi

```python
y = np.loadtxt("chi.txt")
x= np.arange(100)

plt.plot(x, y)
plt.xlabel("j")
plt.ylabel("$\chi^2$")
plt.title("$\chi^2$")
plt.grid(True)
plt.show()
```

Il comportamento oscillante di questo grafico, e i numerosi blocchi aventi $\chi^2>110$ ci confermano che il nostro strumento di generazione di numeri casuali, pur essendo abbastanza sofisticato, non è comunque ideale.
