Stampiamo i risultati di $\pi$ ottenuti simulando un esperimento di Buffon. Come al solito, stampiamo medie e incertezze progressive in funzione del numero di blocchi, usando $N=10^5$ passi e $N_{blocks} =100$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y = np.loadtxt("mean_03.txt")
err = np.loadtxt("err_03.txt")
with open("integers.txt") as file:
    riga = file.readline()
    N, L = map(int, riga.split())

x = np.arange(N)
x *= L

Pi = np.ones(x.size)*math.pi

plt.errorbar(x,y,yerr=err, label = 'Buffon experiment')
plt.plot(x, Pi, color ='r', label = 'True value')
plt.xlabel('#throws')
plt.ylabel('$\pi$')
plt.grid(True)
plt.legend()
plt.show()
```
