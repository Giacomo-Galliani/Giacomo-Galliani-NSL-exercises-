Stampiamo i risultati per i dati standard: 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

#Distribuzione Standard 


x= np.loadtxt("histo11.txt")
y = np.loadtxt("histo21.txt")
bin_edges = np.arange(1, 8, 1)

# Istogrammi con contorno visibile e interni trasparenti
plt.hist(x, bins=bin_edges, facecolor='none', edgecolor='blue', linewidth=1.5)


# Etichette e legenda
plt.title("Distribuzione standard, N=1")
plt.xlabel("Risultati del lancio")
plt.ylabel("Frequenza")

# Mostra il grafico
plt.show()

bin_edges = np.arange(1, 7, 0.5)
plt.hist(y, bins=bin_edges, facecolor='none', edgecolor='red', linewidth=1.5)

# Etichette e legenda
plt.title("Distribuzione standard, N=2")
plt.xlabel("Risultati del lancio")
plt.ylabel("Frequenza")

# Mostra il grafico
plt.show()

bin_edges = np.arange(1, 7, 0.2)
z =np.loadtxt("histo31.txt")

plt.hist(z, bins=bin_edges, facecolor='none', edgecolor='green', linewidth=1.5)

# Etichette e legenda
plt.title("Distribuzione standard, N=10")
plt.xlabel("Risultati del lancio")
plt.ylabel("Frequenza")

plt.show()

bin_edges = np.arange(1, 7, 0.1)
l =np.loadtxt("histo41.txt")

plt.hist(l, bins=bin_edges, facecolor='none', edgecolor='purple', linewidth=1.5)

# Etichette e legenda
plt.title("Distribuzione standard, N=10")
plt.xlabel("Risultati del lancio")
plt.ylabel("Frequenza")

plt.show()
```
