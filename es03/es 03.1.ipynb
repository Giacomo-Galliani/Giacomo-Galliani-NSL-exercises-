Per prima cosa, stampiamo i valori per $C[S(T), T]$, campionando direttamente $S(T)$: (in tutte le simulazioni si sono usati $10^6$ campionamenti, con $N_{blocks} =100$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

#analitical solutions
call = 14.975790778311286 
put = 5.4595325819072364 

#read data from file 
y = np.loadtxt("mean_01.txt")
err = np.loadtxt("err_01.txt")
with open("integers.txt") as file:
    riga = file.readline()
    N, L = map(int, riga.split())

x = np.arange(N)
x *= L

#fills y_call, y_put for comparison 
y_call = np.zeros(x.size)
y_put = np.zeros(x.size)

for i in range(x.size):
    y_call[i] = call 
    y_put[i] = put

#plot results with uncertenties 
plt.errorbar(x,y,yerr=err, label = 'MC simulation')
plt.xlabel('#throws')
plt.ylabel('$C [S(T), T]$')
plt.grid(True)

plt.plot(x, y_call, label= 'analitic solution', color = 'r')
plt.legend()
plt.show()
```

Ora eseguiamo la stessa cosa per $P[S(T), T]$: 

```python
#read data from file
y = np.loadtxt("mean_02.txt")
err = np.loadtxt("err_02.txt")

#plot results 
plt.errorbar(x,y,yerr=err, label = "MC simulation")
plt.xlabel('#throws')
plt.ylabel('$P[S(T), T]$')
plt.grid(True)

plt.plot(x, y_put, label= 'analitic solution', color = 'r')
plt.legend()

plt.show()
```
