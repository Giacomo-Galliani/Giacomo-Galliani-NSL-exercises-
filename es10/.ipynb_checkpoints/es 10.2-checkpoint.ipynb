Per prima cosa stampiamo la traiettoria trovata e la lunghezza del percorso migliore in funzione del numero di generazioni: 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

z = np.loadtxt("sorted_cities.dat")
z.reshape(-1,2)

x =z[:,0]
y =z[:,1]

plt.plot(x,y)
plt.grid(True)
plt.title('Sorted cities')
plt.show()

z = np.loadtxt("lenght.dat")
z = z.reshape(-1,2)

x =z[:,0]
y =z[:,1]



plt.plot(x,y)
plt.grid(True)
plt.title('Lenght over generations')
plt.xlabel('Generations')
plt.ylabel('$L^{(2)}$')
plt.show()
```

La lunghezza finale (in $L^{(2)}$) è dunque data da:

```python
print(y[y.size-1])
```

Come sarebbero stati i nostri risultati senza usare un algoritmo parallelo? Confrontiamo i dati che abbiamo trovato con un algorirmo sempre su 11 popolazioni, che non usi però la migrazione: 

```python
z = np.loadtxt("lenght_comp.dat")
z = z.reshape(-1,2)

y2 =z[:,1]



plt.plot(x,y, label = 'with migration')
plt.grid(True)
plt.title('Lenght over generations (Migration vs indipendent search)')
plt.xlabel('Generations')
plt.ylabel('$L^{(2)}$')
plt.plot(x, y2, label = 'without migration')
plt.legend()
plt.show()
```

e i valori finali, a parità di numero di generazioni sono: 

```python
print(y[y.size-1])
print(y2[y2.size-1])
```

```python
target = y2[y2.size-1]
count =0

for i in range(y.size):
    if(y[i]>target):
        count +=1

print(count*5)
```

Ovvero l'algoritmo parallelo ci mette solo 325 generazioni a "superare" l'algoritmo indipendente. 

In entrambi i casi, si sono usate $1000$ generazioni di $11$ continenti, aventi una popolazioni di $440$ individui.
