Prima di iniziare la simulazione, è necessario assicurarsi che il sistema abbia raggiunto l'equilibrio. La nostra situazione iniziale è quella dove ogni spin è disposto random in up o down, ovvero il limite di $T \rightarrow \infty$.
La nostra simulazione invece parte da una temperatura di $T^* = 0.5$: stampiamo allora i valori dell'energia interna istantanei per determinare quando l'equilibrio è raggiunto. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

z = np.loadtxt("total_energy_beginning.dat")
z.reshape(-1,4)

x = z[:, 0]
y = z[:, 1]

plt.plot(x[:500], y[:500])
plt.grid(True)
plt.show()
```

Vediamo dunque che un numero di circa 200 passi è sufficente per raggiungere l'equilibrio. Eseguiti i 200 passi, possiamo cominciare la nostra simulazione: per prima cosa stampiamo alcuni valori dell'energia interna al variare di $T \in [0.5, 2.5]$, ottenuti usando l'algoritmo di metropolis. 

```python
z1 = np.loadtxt("tenergy_T_dependence.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical Simulation')
plt.grid(True)


points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e, label= 'Analitic solution')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()
plt.show()
```

Il calore specifico a volume costante: 

```python
z1 = np.loadtxt("cv_T_dependence.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical simulation')
plt.grid(True)


heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label = 'Analitic solution')
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.legend()
plt.show()
```

e infine la magnetizzazione: 

```python
z1 = np.loadtxt("magnet_T_dependence.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical simulation')
plt.grid(True)

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M, label= 'Analitic solution')
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()
plt.show()
```

Ora vediamo i risultati delle simulazioni se al posto di usare l'algoritmo di Metropolis usiamo l'algoritmo di Gibbs:

```python
z1 = np.loadtxt("tenergy_T_dependence_Gibbs.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical Simulation')
plt.grid(True)


points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e, label= 'Analitic solution')
plt.title('Ising 1D, internal energy (Gibbs algorithm)')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()
plt.show()


z1 = np.loadtxt("chi_T_dependence_Gibbs.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical simulation')
plt.grid(True)


X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, label = 'Analitic solution')
plt.title('Ising 1D, Susceptibility (Gibbs algorithm)')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.legend()
plt.show()

z1 = np.loadtxt("cv_T_dependence_Gibbs.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical simulation')
plt.grid(True)


heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label = 'Analitic solution')
plt.title('Ising 1D, Heat capacity (Gibbs algorithm)')
plt.xlabel('T')
plt.ylabel('C')
plt.legend()
plt.show()

z1 = np.loadtxt("magnet_T_dependence_Gibbs.dat")
z1.reshape(-1,3)

x1 = z1[:, 0]
y1 = z1[:, 1]
err1 = z1[:,2]

plt.errorbar(x1, y1, yerr =err1, marker='o', linestyle ='None', label = 'Numerical simulation')
plt.grid(True)

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M, label= 'Analitic solution')
plt.title('Ising 1D, magnetization M with h = 0.02 (Gibbs algorithm)')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()
plt.show()
```

Come ci ci aspettavamo, entrambi i risultati ottenuti con i due diversi algoritmi sono compatibili. Ad eccezzione per il calore specifico a volume costante, sono compatibili anche con le soluzioni analitiche.
