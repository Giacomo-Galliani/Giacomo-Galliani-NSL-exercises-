Vediamo ora il comportamento del nostro sistema sotto inversione temporale: 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

z = np.loadtxt("potential_energy.dat")
delta = 0.001
z = z.reshape(-1,4)
y = z[:, 1]
x = np.arange(y.size) *delta


y2 = np.zeros(y.size)
for i in range (int(y.size/2)):
    y2[i] = y[i]
    y2[y.size -1 -i] = y[i]
    

plt.plot(x,y, label = 'full MD simulation')
plt.xlabel('$t$')
plt.ylabel('$U/N$')
plt.plot(x, y2, label = 'Theoretical time inversion')
plt.legend()
plt.grid(True)
plt.show()
```

Nella figura sono riportate le medie dei blocchi (non progressive, se no sarebbe stato difficile vedere le proprietà sotto inversione temporale). L'inversione temporale è effettuata a metà, ovvero a $t=0.04$. Fino a circa $t =0.05$ le misure sono compatibili, poi essendo il sistema a molte particelle caotico, un errore di calcolo piccolo porta a leggermente diverse condizioni iniziali e dunque diverse traiettorie (dalla definizione di caocità) . In questo senso non ci stupisce che la configurazione finale non sia uguale alla iniziale.
