Per iniziare, stampiamo i dati ottenuti con 34 città distribuite random su un cerchio, il percorso ottimale è dato da: 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

z = np.loadtxt("sorted_cities_0.dat")
z.reshape(-1,2)

x =z[:,0]
y =z[:,1]

plt.title('Best trajectory (circumference)')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

Che come ci aspettavamo, rappresenta il percorrere il cerchio in senso orario o anti-orario (in questo caso anti-orario).

Stampiamo anche la lunghezza del percorso e della media della metà dei percorsi più corti, in funzione del numero di generazioni: 

```python
z = np.loadtxt('lenght_0.dat')
z = z.reshape(-1, 3)
x = z[:, 0]
y1 = z[:, 1]
y2 = z[:,2]


plt.plot(x, y1, label = 'best path')
plt.plot(x, y2, label = 'mean of half best paths')
plt.legend()
plt.xlabel('Generations')
plt.ylabel('$L^{(2)}$')
plt.grid(True)
plt.show()
```

Ora invece stampiamo i risultati per le città generate random all'interno di un quadrato: 

```python
z = np.loadtxt("sorted_cities_1.dat")
z.reshape(-1,2)

x =z[:,0]
y =z[:,1]

plt.title('Best trajectory (square)')
plt.plot(x,y)
plt.grid(True)
plt.show()

z = np.loadtxt('lenght_1.dat')
z = z.reshape(-1, 3)
x = z[:, 0]
y1 = z[:, 1]
y2 = z[:,2]


plt.plot(x, y1, label = 'best path')
plt.plot(x, y2, label = 'mean of half best paths')
plt.legend()
plt.xlabel('Generations')
plt.ylabel('$L^{(2)}$')
plt.grid(True)
plt.show()
```
