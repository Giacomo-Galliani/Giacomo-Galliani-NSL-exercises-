Per prima cosa stampiamo l'energia all'abbassarsi della temperatura: 

```python
import matplotlib.pyplot as plt 
import numpy as np

z = np.loadtxt("energy.dat")
z = z.reshape(-1, 5)
x = z[:, 0]
y = z[:, 1]
err = z[:, 2]

plt.errorbar(x, y, yerr = err)
plt.xlabel("1/T")
plt.ylabel("Energy")
plt.grid(True)
plt.show()
```

Una volta trovato il presunto stato fondamentale, possiamo campionare con l'algoritmo di metropolis la distribuzione $|\psi_{GS}(x)|^2$, e confrontarla con il risultato ottenuto tramite tecniche di algebra lineare: 

```python
#load data form file 
z = np.loadtxt("histo.dat")

#plot data
plt.hist(z, 100)


def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

#plot solution 
plt.plot(x,57000*(psi[0])**2)
plt.show()
```

Ora possiamo visualizzare la traiettoria nello spazio dei parametri: 


```python
z = np.loadtxt("parameters.dat")
z = z.reshape(-1, 2)
x = z[:, 0]
y = z[:, 1]

plt.plot(x, y)
plt.xlabel("$\mu$")
plt.ylabel("$\sigma$")
plt.grid(True)
plt.show()
```

Non avendogli detto che il parametro $\mu$ debba necessariamente essere positivo, il problema aveva due minimi globali. Nella prima fase l'algoritmo stava convergendo verso un $\mu$ positivo, per poi saltare nel minimo con $\mu < 0$.

Infine diamo il valore medio dell'hamiltioniana e la sua incertezza in funzione del numero di blocchi: 

```python
z = np.loadtxt("pro_energy.dat")
z = z.reshape(-1, 3)
x = z[:, 0]
y = z[:, 1]
err = z[:, 2]
e = -0.4604658796925906

gs = np.ones(x.size)*e

plt.errorbar(x, y, yerr = err, label = 'Simulated Anniling')
plt.plot(x, gs, label = 'Algebraic solution', color = 'r')
plt.xlabel("blocks")
plt.ylabel("Energy")
plt.grid(True)
plt.legend()
plt.show()
```

Il risultato ottenuto con il metodo del simulated anniling non è compatibile con il valore del GS trovato traminte tecniche di algebra lineare. Questo significa che probabilmente l'algoritmo non ha raggiunto pienamente il vero stato fondamentale.
