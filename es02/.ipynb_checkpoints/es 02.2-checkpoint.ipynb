Stampiamo la radice della distanza quadratica media in un random walk prima per il caso reticolare, poi per il caso continuo, in funzione del numero di passi. Ci aspettiamo che la funzione segua una legge proporzionale alla radice quadrata del numero di passi (trattandosi di un processo diffusivo). 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y = np.loadtxt("mean_03.txt")
err = np.loadtxt("err_03.txt")
with open("integers.txt") as file:
    riga = file.readline()
    N, L = map(int, riga.split())

x = np.arange(N)
x *= L

plt.errorbar(x,y,yerr=err)
plt.xlabel('#passi')
plt.ylabel('radice della distanza quadratica media')
plt.grid(True)
plt.show()
```

Osservo che la curva sembra essere una funzione del tipo $f(N) = kN^{1/2}$. Per esserne sicuro plotto anche la funzione al quadrato:

```python
y = np.loadtxt("mean_04.txt")
err = np.loadtxt("err_04.txt")

linear_fit = np.polyfit(x, y, 1)

plt.errorbar(x,y,yerr=err, label = 'simulation data')
plt.xlabel('#passi')
plt.ylabel('distanza quadratica media')
plt.grid(True)
plt.plot(x, linear_fit[0]*x + linear_fit[1], label = 'linear fit ')
plt.legend()
plt.show()

chi2 =0

for i in range(x.size):
    if err[i] != 0:
        chi2 += ((linear_fit[0]*x[i] + linear_fit[1] - y[i])/err[i])*((linear_fit[0]*x[i] + linear_fit[1] - y[i])/err[i])
        
print("chi^2 =", chi2)
```

Statisticamente dunque, concludiamo che la distanza quadratica media è compatibile con una legge $\propto N^{1/2}$ (poichè $\chi^2 = 217 << N=10^6$)
