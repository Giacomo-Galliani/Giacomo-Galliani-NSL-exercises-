Per prima cosa, stampiamo i risultati dell'integrale $\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$ ottenuti campionando una distribuzione uniforme, campionando $10^6$ volte l'intervallo $[0, 1)$, con $N_{blocks}=100$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y = np.loadtxt("mean_01.txt")
err = np.loadtxt("err_01.txt")
with open("integers.txt") as file:
    riga = file.readline()
    N, L = map(int, riga.split())

x = np.arange(N)
x *= L

plt.errorbar(x,y-1,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<r>-1')
plt.grid(True)
plt.show()
```

Ora stampiamo i risultati ottenuti usando l'importance sampling: per farlo abbiamo preso lo sviluppo di Taylor al primo ordine di $f(x) = \frac{\pi}{2}\cos(\pi x/2)$, centrata in $x=1$. La scelta del punto è stata determinata dal fatto che tale sviluppo è positivo per $x \in [0,1]$. L'altra prerogativa per essere una distribuzione valida è che sia correttamente normalizzata, così che la nostra distribuzione diventa $p(x) = 2(1-x)$ (e la funzione da valutare $g(x) = \frac{\pi}{4(1-x)}\cos(\pi x/2)$).
